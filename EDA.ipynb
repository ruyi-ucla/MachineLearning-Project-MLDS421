```python
from ucimlrepo import fetch_ucirepo 
from ucimlrepo import fetch_ucirepo
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
# fetch dataset 
secondary_mushroom = fetch_ucirepo(id=848) 
  
# data (as pandas dataframes) 
X = secondary_mushroom.data.features 
y = secondary_mushroom.data.targets 

df = pd.concat([X, y], axis=1)
# metadata 
print(secondary_mushroom.metadata) 
  
# variable information 
print(secondary_mushroom.variables) 
```

```python
categorical_features = X.select_dtypes(include=['object']).columns

# Adjusting the subplot layout for 17 categorical features to be visually appealing
num_features = 17  # Given number of categorical features
rows = 4  # Setting 4 rows for better spacing
cols = (num_features + rows - 1) // rows  # Distribute evenly across rows

# Create subplots
fig, axes = plt.subplots(rows, cols, figsize=(20, 16))

# Flatten axes for easy iteration
axes = axes.flatten()

# Generate count plots for each categorical feature
for i, col in enumerate(categorical_features[:num_features]):  # Ensure no indexing error
    sns.countplot(y=X[col], order=X[col].value_counts().index, palette='coolwarm', ax=axes[i])
    axes[i].set_title(f'Distribution of {col}', fontsize=14)
    axes[i].set_xlabel("Count")
    axes[i].set_ylabel(col)

# Remove any empty subplots (if less than total grid cells)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout for better visibility
plt.tight_layout()
plt.show()

```

### Checking correlation (only applicable to numerical features)

```python
numerical_features = X.select_dtypes(include=['number']).columns
if len(numerical_features) > 0:
    plt.figure(figsize=(10, 6))
    sns.heatmap(X[numerical_features].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.show()
else:
    print("No numerical features found for correlation analysis.")
```

### Encoding categorical variables for correlation analysis

```python
df_encoded = pd.get_dummies(X, drop_first=True)
plt.figure(figsize=(12, 8))
sns.heatmap(df_encoded.corr(), cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Encoded Features')
plt.show()
```
